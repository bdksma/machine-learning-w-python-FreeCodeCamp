```python
import tensorflow as tf

string = tf.Variable("this string", tf.string)
int = tf.Variable(362, tf.int16)
float = tf.Variable(4.05, tf.float16)

print(string)
print(int)
```

Rank/Degrees in Tf


```python
rank_1degree = tf.Variable(["This", "is", "One"], tf.string)
rank_2degree = tf.Variable([["This", "is", "One","yes"], ["this", "is", "two", "no"]], tf.string)

tf.rank(rank_2degree)
```
